### Ejercicio 1

Implementar una Pila utilizando la siguiente definición de una *Lista 'Simple' Enlazada*. Utilizar el prototipo de la clase Lista Simple Enlazada mostrado a continuación.

```python
class Pila_Enlazada:
    """Clase Lista 'Simplemente' Enlazada."""

#-------- Clase Anidada - NODO ------------#
    class _Node:
        """Clase Nodo."""
        __slots__ = '_element', '_next' # optimiza el uso de memoria

        def __init__(self, element, next):
            self._element = element # inicializar el contenido del Nodo
            self._next = next       #referencia al siguiente Nodo

    def __init__(self):
        """Crea una Pila Vacia."""
        self._head = None  #referencia a la "cabeza" de la lista
        self._size = 0     #cantidad de elementos
    ###

    def __len__(self):
        """Retorna el numero de elementos de la Pila."""
        return self._size

    def is_empty(self):
        """Retorna VERDADERO si la Pila esta Vacia."""
        return self._size == 0

    #----------- Métodos de la Pila ----------- #

    def push(self, elem): # Esto sirve para comparar con el anterior y se va ordenando de a poco
        new_node = self._Node(elem, self._head)
        self._head = new_node
        self._size +=1

    def pop(self):
        if self.is_empty():
            raise IndexError('La pila esta vacia')
        elem = self._head._element
        self._head = self._head._next
        self._size -= 1
        return elem
    
    def top(self):
        if self.is_empty():
            raise IndexError('La pila esta vacia')
        return self._head._element
    
    def peek(self):
        return self.top()

```

```python
# Crear una pila vacía
pila = Pila_Enlazada()

# Verificar si está vacía
print("¿La pila está vacía?", pila.is_empty())  # True

# Agregar elementos
pila.push(10)
pila.push(20)
pila.push(30)

# Ver el tamaño
print("Tamaño de la pila:", len(pila))  # 3

# Ver el elemento del tope sin quitarlo
print("Elemento en el tope (peek):", pila.peek())  # 30

# Sacar elementos (pop)
print("Elemento retirado:", pila.pop())  # 30
print("Elemento retirado:", pila.pop())  # 20

# Ver el tope después de algunos pop
print("Elemento en el tope ahora:", pila.top())  # 10

# Ver si la pila está vacía
print("¿La pila está vacía?", pila.is_empty())  # False

# Sacar el último elemento
print("Elemento retirado:", pila.pop())  # 10
print("¿La pila está vacía?", pila.is_empty())  # True

# Intentar sacar un elemento de una pila vacía (esto dará error)
# pila.pop()  # Descomenta para probar la excepción
```

Una cola del súper es FIFO (First In, First Out): el primero que llega es el primero que se atiende.

Tu pila es LIFO (Last In, First Out): el último que llega es el primero que se retira, como una torre de platos.

```python
class ColaEnlazada:
"""FIFO queue implementation using a singly linked list for storage."""

#-------- Clase Anidada - NODO ------------#
  class _Node:
    """Clase Nodo."""
    __slots__ = '_element', '_next' # optimiza el uso de memoria

    def __init__(self, element, next):
      self._element = element # inicializar el contenido del Nodo
      self._next = next       #referencia al siguiente Nodo

#----------- Métodos de la Cola ----------- #

  def __init__(self):
    """Crea una Cola Vacia."""
    self._head = None
    self._tail = None
    self._size = 0  # Numero de elementos en la Cola

  def __len__(self):
    """Retorna el numero de elementos de la Cola."""
    return self._size

  def is_empty(self):
    """True si la Cola esta vacia."""
    # COMPLETAR #
    return self._size == 0

  def first(self):
    """Retornar (sin remover) el primer elemento de la Cola.

       Raise Empty exception if the queue is empty. """
    # COMPLETAR # 
    if self.is_empty():
        raise Empty('Queue is empty')
    else:
        # Retorna el valor del primer nodo
        return self._head._element

  def dequeue(self):
    """Remove and return the first element of the queue (i.e., FIFO).

       Raise Empty exception if the queue is empty."""
        # Ejemplo:   ## if self.is empty():
        #            ##    raise Empty( Queue is empty )
    if self.is_empty():
        raise Empty('Queue is empty')
    element = self._head._element # Guardar elemento a devolver
    self._head = self._head._next # Muevo la cabeza al siguiente elemento
    self._size -= 1
    if self.is_empty():             # si quedó vacía, actualizar cola
      self._tail = None
      return element

       
  def enqueue(self, e):
    """Add an element to the back of queue."""
    # COMPLETAR #
    new_node = self._Node(e, None)       # Crear un nuevo nodo con el valor e
    if self.is_empty():                   # Si la cola está vacía
        self._head = new_node             # El nuevo nodo es la cabeza
        self._tail = new_node             # Y también la cola
    else:
        self._tail._next = new_node       # Conectar el nuevo nodo al final
        self._tail = new_node             # Actualizar la cola al nuevo nodo
    self._size += 1                       # Aumentar el tamaño de la cola
```

 Diferencia con una pila

Cola (FIFO): el primero en entrar es el primero en salir.

Operaciones: enqueue (agregar al final), dequeue (sacar del frente)

Pila (LIFO): el último en entrar es el primero en salir.

Operaciones: push (agregar arriba), pop (sacar del tope)

```python
def __str__(self):
    elementos = []
    actual = self._head
    while actual is not None:
        elementos.append(actual._element)
        actual = actual._next
    return f"Cola: {elementos}"


def __add__(self, otro):
    nueva_cola = ColaEnlazada()
    actual = self._head
    while actual is not None:
        nueva_cola.enqueue(actual._element)
        actual = actual._next
    actual = otro._head
    while actual is not None:
        nueva_cola.enqueue(actual._element)
        actual = actual._next
    return nueva_cola


```
